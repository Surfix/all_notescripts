```python
#read_csv
SWE_2020 = pd.read_csv('/home/naheemadebisi/PhD/snow-analytics/Radar-Backscatter/pits_data/2020/SNEX20_TS_SP_preliminary_v6/SNEX20_TS_SP_Summary_SWE_v01.csv')
#create column for mean temp
SWE_2020['Temp_mean'] = np.nan
```

```python


#list all site directories
site_list = glob.glob('/home/naheemadebisi/PhD/snow-analytics/Radar-Backscatter/pits_data/2020/SNEX20_TS_SP_preliminary_v6/pits/*/*')

for site_dir in site_list:
    if os.path.isdir(site_dir):
        #extract pidID
        pitID = os.path.basename(site_dir)
        #check if the pitID exist in dataframe
        if pitID in SWE_2020['PitID'].values:
            #print(f'{site_dir} is found')
            #print(pitID)

            #list all csv file, select the temperature csv
            all_files = glob.glob(os.path.join(site_dir, "*.csv"))
            for file in all_files:
                if 'temperature' in file:
                    #print(file)
                    temp = pd.read_csv(file, skiprows=10)
            #find mean
            SWE_2020.loc[SWE_2020['PitID'] == pitID, 'Temp_mean'] = temp['Temperature (deg C)'].mean()
            #SWE_2020.loc[pitID, 'Temp_mean'] = temp['Temperature (deg C)'].mean()
```
