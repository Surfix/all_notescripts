```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
density = np.linspace(200, 300)

perm = 1 + 1.6 * (density/1000) + 1.8 * (density/1000)**3
```

```python
np.median(density)
```

```python
def phase_change(sd, inc_deg, permitivity):
    inc = (inc_deg * np.pi)/180
    delta_phase = (4* np.pi * sd *(np.cos(inc) - np.sqrt(permitivity - np.sin(inc)**2)) )/0.25
    print(f'range of phase change is {np.max(delta_phase) - np.min(delta_phase)}')
    print(f'median of permittiviy {np.median(permitivity)}')
    return delta_phase
```

```python
que1 = phase_change(1, 40, perm)
```

```python
# Set font size and font family of matplotlib for plotting
plt.rcParams.update({'font.size': 14})
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']

fig, axs = plt.subplots(nrows = 1, ncols = 2, figsize=(16,11))
sns.scatterplot(density, perm, ax = axs[0])
axs[0].axvline(np.median(density))
axs[0].axhline(np.median(perm), color = 'r')
axs[0].set_ylabel('permitivity')
axs[0].set_xlabel('density')


sns.scatterplot(density, que1, ax = axs[1])
axs[1].set_ylabel('phase change')
axs[1].set_xlabel('density')
axs[1].axvline(np.median(density))
axs[1].axhline(np.median(que1), color = 'r')
```
